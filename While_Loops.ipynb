## What are loops?
- A loop is a control structure that is used to perform a set of instructions for a specific number of times.

##### Loops are a programming element that repeat a portion of code a set number of times until the desired process is complete. Repetitive tasks are common in programming, and loops are essential to save time and minimize errors.

## While loop
- The while loop keeps iterating over a certain set of operations as long as a certain condition holds True.


```python
# Q1 : Print from 1 to 10

num = 1
while num <= 10 :
  print(num)
  num = num + 1
```

```python
# Q2 : Take a value N as input and print from 1 to N

# N = 3
# 1
# 2
# 3

n = int(input())

count = 1
while count <= n :
  print(count)
  count = count + 1
```

```python
# Quiz 3 - infinite loop
count = 1
while count <= 10:
  print(10)

# Output
# 10
# 10
# 10
```

```python
# Q3 : Take a value N as input and print from N to 1

n = int(input())

count = n
while count >= 1 :
  print(count)
  count -= 1
```

```python
# Q4 : Take a value N as input and print odd numbers from 1 to N

# N = 10
# Output : 1 3 5 7 9

n = int(input("Give the value of N: "))

# Approach 1 - Go over all values, print only odd
print("Approach 1")
count = 1
while count <= n :
  if count % 2 == 1 :
    print(count)
  count += 1

# Approach 2 - Go over only odd values and print
print("Approach 2")
count = 1
while count <= n :
  print(count)
  count += 2
```

```python
# Q6 : Take a value N as input and print the sum of even numbers till N

# N = 11
# Output : 2+4+6+8+10 = 30

n = int(input("Give N as input: "))

sum = 0
count = 2
while count <= n :
  sum = sum + count
  count = count + 2

print(sum)

# sum = 0
# sum = sum + 2 = 0 + 2 = 2
# sum = sum + 4 = 2 + 4 = 6
# sum = sum + 6 = 6 + 6 = 12
# sum = sum + 8 = 12 + 8 = 20
# sum = sum + 10 = 20 + 10 = 30
```

```python
# Q7 : Take a value N as input. Then take N integers as input. Print their sum.
# Input
# 3 -> N
# -2 -> V1
# 1 -> V2
# 5 -> V3
# Sum = -2 + 1 + 5 = 4

n = int(input("Give N as input: "))
count = 1
sum = 0
while count <= n :
  val = int(input("Give input: "))
  sum = sum + val
  count = count + 1

print("Sum is", sum)
```
